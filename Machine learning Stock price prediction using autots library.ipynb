```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
data = pd.read_csv("C:\Users\swappy\Downloads\TATASTEEL.NS")
print(data.head())

```

```python
pip install autots

#AutoTS means Automatic Time Series, it is a machine learning library in Python used for the task of time series forecasting.
#I recently used it for forecasting the stock prices of Apple for the next 10 days and the resulting figures were so accurate
```

```python
#visualize an interactive visualisation of the stock prices

figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"], high=data["High"],
                                        low=data["Low"], close=data["Close"])])
figure.update_layout(title = "Tata Steel Stock Price Analysis", xaxis_rangeslider_visible=False)
figure.show()
```

```python
#correlation between the features of this dataset
print(data.corr())
```

```python
#using the autots library in Python
#If you have never used this Python library before, then you can easily install it by using the pip command
# pip install autots

#predict the stock prices
from autots import AutoTS
model = AutoTS(forecast_length=5, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
